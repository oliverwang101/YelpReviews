# Exploration

## Get sentiment per review using nltk.sentiment.vader

```python
# read in full reviews from kaggle data, clean up reviews (lowercase, remove all non-letters)
# remove all stop words from nltk library
```

```python
import pandas as pd

import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

revs = pd.read_csv("full_reviews.csv").drop("Unnamed: 0",axis=1)

revs["words"] = revs["text"].str.lower().str.replace("[^A-Za-z\s]", " ")
stop_words = set(stopwords.words("english"))
```

```python
def remove_stop_words(sentence):
    words = word_tokenize(sentence)
    filtered_revs = []
    for w in words:
        if w not in stop_words:
            filtered_revs.append(w)
    new_revs= ' '.join(filtered_revs)
    return new_revs
```

```python
def get_sentiments(revs):    
    sid = SentimentIntensityAnalyzer()
    sss = {"neg":[],"neu":[],"pos":[]}
    for index,rev in enumerate(revs):
        ss = sid.polarity_scores(rev)
        sss["neg"].append(ss["neg"])
        sss["neu"].append(ss["neu"])
        sss["pos"].append(ss["pos"])
    return sss
```

## All reviews mean sentiment (neu, pos, neg) by rating

```python
revs["words"] = revs["words"].apply(remove_stop_words)
```

```python
sss = get_sentiments(revs["words"])
```

```python
revs = pd.concat([pd.DataFrame(sss),revs],axis=1)
```

```python
revs["rating"] = revs.rating.astype(str)
```

```python
revs.groupby("rating")[["neu","neg","pos"]].mean().plot.bar()
```

# Negative sentiment proportion distribution by Rating

```python
import numpy as np
ratings = revs["rating"].unique().astype(int)
ratings = np.sort(ratings).astype(str)
```

```python
fig, axes = plt.subplots(1, 5 , figsize=(20, 3), sharex=True, sharey=True)

for ax, rating in zip(axes, ratings):
    rat_type = revs[revs["rating"] == rating]
    rat_type.neg.plot.hist(ax=ax)
    ax.set_title("Neg Distribution for Rating %s"%(rating))
```

# Positive sentiment proportion distribution by Rating


```python
ratings = revs["rating"].unique().astype(int)
ratings = np.sort(ratings).astype(str)
fig, axes = plt.subplots(1,5, figsize=(20, 3), sharex=True, sharey = True)
for ax, rating in zip(axes, ratings):
    rat_type = revs[revs["rating"] == rating]
    rat_type.pos.plot.hist(ax=ax)
    ax.set_title("Pos Distribution for Rating %s"%(rating))
    
```

```python
revs["rest_rating"] = revs["rest_rating"].astype(str)
revs.groupby(["rest_rating"]).review_count.mean().plot.bar()
plt.ylabel('Review Count')
plt.xlabel('Restaurant Rating')
plt.title('Review Count by Restaurant Rating')
```

```python
revs.groupby("rating").neg.count().plot.bar()
plt.xlabel('Review Rating')
plt.ylabel('Count')
plt.title('Number of Reviews by Review Rating')
```

# Distributions of sentiment (pos, neg) by rating and "business type"

```python
revs.categories = revs.categories.str.replace(" ","")
revs.categories
revs["service"] = revs.categories.fillna("").str.split(",")
```

```python
%matplotlib inline
from sklearn.preprocessing import MultiLabelBinarizer

binarizer = MultiLabelBinarizer()
services = pd.DataFrame(
    binarizer.fit_transform(revs["service"]),
    columns=binarizer.classes_
)
```

```python
services = services.drop('',axis=1)
```

```python
revs_services = pd.concat([revs,services],axis=1)
```

```python
serv = revs_services.columns[15:]
ordered_serv = revs_services[serv].sum(axis=0).sort_values(ascending=False)[1:]
```

```python
revs_services.rest_rating = revs_services.rest_rating.astype(str)
```

```python
fig, axes = plt.subplots(5,3, figsize=(25,40), sharey=True)

for ax, serv in zip(axes.flatten(), ordered_serv.index[:20]):
    service = revs_services[revs_services[serv]==1]
    service.groupby("rest_rating")[["neg","pos"]].mean().plot.bar(ax=ax)
    ax.set_title("Sentiment by Rating & Restaurant(%s)"%(serv))
    ax.set_xlabel('Restaurant Rating')
    ax.set_ylabel('Proportion')
```
